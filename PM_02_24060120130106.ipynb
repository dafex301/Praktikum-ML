```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling
import numpy as np
```

```python
# Create x and y coordinates
x = np.array([0.40, 0.22, 0.35, 0.26, 0.08, 0.45])
y = np.array([0.53, 0.38, 0.32, 0.19, 0.41, 0.30])
```

```python
# Plot the points
plt.scatter(x, y, s=100);
```

```python
# Set 3 initial centroids
centroids = np.array([[0.2, 0.4],
                        [0.4, 0.2],
                        [0.4, 0.4]])

# Plot the centroids with 3 different colors
plt.scatter(x, y, s=100)
plt.scatter(centroids[:, 0], centroids[:, 1], c='r', s=200);
```

```python
# Calculate the distance between each point and each centroid
distances = np.zeros((6, 3))
for i in range(6):
    for j in range(3):
        distances[i, j] = np.sqrt((x[i] - centroids[j, 0]) ** 2 + (y[i] - centroids[j, 1]) ** 2)

distances
```

```python
# Assign each point to the closest centroid
labels = np.argmin(distances, axis=1)
labels
```

```python
# Plot the points and the centroids and show the color of each cluster
plt.scatter(x, y, c=labels, s=100, cmap='viridis')
plt.scatter(centroids[0][0], centroids[0][1], c='purple', s=200);
plt.scatter(centroids[1][0], centroids[1][1], c='green', s=200);
plt.scatter(centroids[2][0], centroids[2][1], c='orange', s=200);
```

```python
# Update the centroids to the mean of the points in each cluster
centroids = np.array([[x[labels == 0].mean(), y[labels == 0].mean()], [x[labels == 1].mean(), y[labels == 1].mean()], [x[labels == 2].mean(), y[labels == 2].mean()]])
centroids
```

```python
# Plot the points and the new centroids
plt.scatter(x, y, c=labels, s=100, cmap='viridis')
plt.scatter(centroids[0][0], centroids[0][1], c='purple', s=200);
plt.scatter(centroids[1][0], centroids[1][1], c='green', s=200);
plt.scatter(centroids[2][0], centroids[2][1], c='orange', s=200);
```
